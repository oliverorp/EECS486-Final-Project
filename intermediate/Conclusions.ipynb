```python
import pandas as pd
```

```python
data = pd.read_csv("classifierInput.csv")
df = pd.DataFrame(data)
```

```python
times = list(df["TimeOfDay"].unique())
reasons = list(df["WatchReason"].unique())
activities = list(df["CurrentActivity"].unique())
productivities = list(df["BinnedProdLoss"].unique())

print(times)
print(reasons)
print(activities)
print(productivities)
```

```python
tupleDict = {}
for time in times:
    for reason in reasons:
        for activity in activities:
            for productivity in productivities:
                tuple = (time, reason, activity, productivity)
                tupleDict[tuple] = 0
            
df["Tuple"] = list(zip(df["TimeOfDay"], df["WatchReason"], df["CurrentActivity"], df["BinnedProdLoss"]))
dftupleDict = df["Tuple"].values
for tup in dftupleDict:
    tupleDict[tup] += 1

sortedtupleDict = sorted(tupleDict.keys(), key = lambda x: tupleDict[x], reverse=True)
```
